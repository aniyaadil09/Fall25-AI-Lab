```python
class Graph:
    def __init__(self, graph):
        self.graph = graph

    def neighbors(self, node):
        return self.graph.get(node, [])

    def h(self, node):
        return 1

    def a_star(self, start, goal):
        open_set = {start}
        closed_set = set()
        g_score = {start: 0}
        parent = {start: None}

        while open_set:
            # Pick node with lowest f = g + h
            current = min(open_set, key=lambda n: g_score[n] + self.h(n))

            if current == goal:
                # Reconstruct path
                path = []
                while current:
                    path.append(current)
                    current = parent[current]
                path.reverse()
                print("Path found:", path)
                return path

            open_set.remove(current)
            closed_set.add(current)

            for neighbor, cost in self.neighbors(current):
                if neighbor in closed_set:
                    continue

                tentative_g = g_score[current] + cost

                if neighbor not in open_set or tentative_g < g_score.get(neighbor, float('inf')):
                    g_score[neighbor] = tentative_g
                    parent[neighbor] = current
                    open_set.add(neighbor)

        print("No path found!")
        return None



graph_data = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)]
}

g = Graph(graph_data)
g.a_star('A', 'D')
```
