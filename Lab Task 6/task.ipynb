```python
def bfs_simple(graph, start):
    visited = []
    queue = [start]

    while queue:
        node = queue[0]        
        queue = queue[1:]      

        if node not in visited:
            visited.append(node)
            queue += graph[node] 
    print("BFS :", visited)

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

bfs_simple(graph, 'A')
```

```python
def bfs_with_queue_tree(tree, root):
    queue = [root]
    visited = []

    while queue:
        current = queue.pop(0)
        visited.append(current)

        left = tree[current][0]
        right = tree[current][1]

        if left is not None:
            queue.append(left)
        if right is not None:
            queue.append(right)

    print("BFS :", visited)



tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': [None, 'F'],
    'D': [None, None],
    'E': [None, None],
    'F': [None, None]
}

bfs_with_queue_tree(tree, 'A')
```
