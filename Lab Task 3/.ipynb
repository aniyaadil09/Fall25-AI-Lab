```python
class Model_BasedReflexAgent :
    def __init__(self):
        self.threshold_temp = 22
        self.heater_temp ="OFF"
    def sensor(self,temp):
        self.current_temp = temp
    def performance(self):
        if self.current_temp <   self.threshold_temp:
            if self.heater_temp != "ON":
                self.heater_temp = "ON"
                return "Heater turned on."
            else:
                return "Heater remain on"
        else:
            if self.heater_temp != "OFF":
                self.heater_temp = "OFF"
                return "Heater turned off."
            else:
                return "Heater remain off"
    
    def actuator (self):
        action = self.performance()
        print(f"{self.current_temp} ______{action}")
        return action

rooms ={}
with open("data.txt" , "r") as file:
    for line in file:
        if line.strip():
            name, temp = line.strip().split(",")
            rooms[name.strip()] = int(temp.strip())

agent = Model_BasedReflexAgent()
with open("output.txt","w") as output_file:
    for room, temp in rooms.items():
        output_file.write(f"{room} :\t")
        agent.sensor(temp)
        result = agent.actuator()
        output_file.write(result + "\n")


```
